#### **Imports**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/clean_penguins.csv')
df.sample(5)
```

##### **Data Visualization: Pairplot and Correlation Heatmap**

```python
numeric_cols = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']
sns.pairplot(df[numeric_cols + ['species']], hue='species')
plt.suptitle('Pairplot of Penuins Mesurements by Species', y=1.02)
plt.show()
```

##### **Correlation Heatmap - if there are four or more numeric columns**

```python
numeric_df = df.select_dtypes(include=[np.number])
if numeric_df.shape[1] >=4:
    plt.figure(figsize=(8,6))
    corr = numeric_df.corr()
    sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
    plt.title('Correlation Heatmap of Numerisc Features')
    plt.show()
else:
    print('Not enough numeric columns to produce a correlation heatmap.')
```

##### **Visualization Plot - Species Count**

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='species', hue='species', palette='Set3', legend=True)
plt.title('Distribution of Penguin Species')
plt.grid()
plt.show()
```

##### **Boxplot - Flipper Length Across Species**

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='species', y='flipper_length_mm', hue='species', palette='Pastel1', legend=True)
plt.title('Flipper Length by Species')
plt.grid()
plt.show()
```

##### **Violin Plot - Body Mass Across Species**

```python
plt.figure(figsize=(8,5))
sns.violinplot(data=df, x='species', y='body_mass_g', hue='species', palette='Pastel2', legend=True)
plt.title('Body Mass by Species')
plt.grid()
plt.show()
```
